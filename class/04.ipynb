```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame({
    "x": [1, 2, 3, 4, 5],
    "y": [66, 44, 39, 2, 1],
    "z": [244, 29, 35, 3, 33]
})

# Correlation matrix
corr = data.corr()

# Heatmap
sns.heatmap(corr, 
            annot=True,
            cmap="rainbow",
            
)
plt.title("Correlation Heatmap")
plt.show()
```
