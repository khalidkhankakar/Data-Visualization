```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python

# Sample data
categories = ['A', 'B', 'C', 'D']
values = [4, 33, 25, 65]

# bar_colors = ['tab:purple', 'tab:cyan', 'tab:yellow', 'tab:black']

# Create a bar chart
plt.bar(categories, values, alpha=1, edgecolor='black')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart")
plt.show()

```

```python
# component bar chart
categories = ['A', 'B', 'C', 'D']
val1 = [41, 33, 25, 19]
val2 = [23, 12, 14,18]
val3 = [14, 32, 12, 33]
val4 = [20, 17,12, 23]

x = np.arange(len(categories))
width = 0.1

plt.bar(x - width, val1, width, label = 'Value 1', color='red')
plt.bar(x , val2, width, label = 'Value 1', color='blue')
plt.bar(x+ width, val3, width, label = 'Value 1', color='purple')
plt.bar(x + width + 0.1, val4, width, label = 'Value 1', color='green')


plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Multiple bar chart")
plt.show()
```

```python
# component bar chart
categories = ['A', 'B', 'C', 'D']
val1 = [32, 33, 25, 35]
val2 = [17, 16, 13, 18]
val3 = [7, 6, 6, 6]
val4 = [3, 2, 3, 2]
width=0.6

plt.bar(categories, val1, width, label = 'Value 1', color='red')
plt.bar(categories, val2, width, label = 'Value 2', color='blue')
plt.bar(categories, val3, width, label = 'Value 3', color='cyan')
plt.bar(categories, val4, width, label = 'Value 4', color='yellow')

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Component bar chart")
plt.show()
```
