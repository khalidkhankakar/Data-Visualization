```python
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
data = pd.DataFrame({
    "x": [1, 2, 3, 4, 5],
    "y": [66, 44, 39, 2, 1],
    "z": [44, 29, 35, 3, 33],
    "u": [53, 40, 43, 2, 44],
})

# Correlation matrix
corr = data.corr()

plt.figure(figsize=(8, 8))
# Heatmap
sns.heatmap(corr, 
            annot=True,
            center=0,
            linewidths=0.1,
            cmap="coolwarm",
            fmt="0.3f"
            )
plt.title("Correlation Heatmap")
plt.show()
```

```python
data = np.random.normal(loc=50, scale=10, size=1000)

plt.hist(data, bins=30, color='blue', edgecolor='red' )
plt.title('Histplot plot')
plt.show
plt.show()
```

```python
sns.histplot(data, bins=30, kde=True)
plt.title('hist plot with kde')

plt.show()
```

```python
sns.violinplot(data=data)
plt.title('vilion plot')
plt.show
```

```python
sns.ecdfplot(data=data)
plt.title('ecdf plot')
plt.show
```

```python
# shade=True
sns.kdeplot(data=data,  color='red',fill=True)
plt.title('kde plot')
plt.show()
```
