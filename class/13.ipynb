```python
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
import pandas as pd
```

```python

data = [7, 9, 10, 15, 18, 21, 22, 23, 25, 29, 90]

plt.boxplot(data, vert=False)
plt.title("Box Plot with Matplotlib")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

```python
#Score of students of ClassA,ClassB,ClassC

class_a = [78, 82, 85, 90, 95, 98, 100]
class_b = [55, 60, 65, 70, 75, 80, 85]
class_c = [88, 85, 89, 90, 93, 97, 99]

data = [class_a, class_b, class_c]
labels = ['Class A', 'Class B', 'Class C']

plt.boxplot(data,labels=labels, notch=0.5, patch_artist=True,
                  boxprops=dict(facecolor='red'))
plt.title("Multiple Box Plots with Matplotlib")
plt.ylabel("Scores")
plt.grid(True)
plt.show()
```

```python
data = [7, 9, 10, 15, 18, 21, 22, 23, 25, 29, 90]

sns.boxplot(data=data, color='yellow')
plt.title("Box Plot with Seaborn")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

```python
Class_A= [78, 82, 85, 90, 95, 98, 100]
Class_B= [55, 60, 65, 70, 75, 80, 85]
Class_C= [88, 85, 89, 90, 93, 97, 99]
df = [Class_A, Class_B, Class_C]

# print(sns.palettes.SEABORN_PALETTES.keys())
sns.boxplot(df, palette='dark')
plt.title("Multiple Box Plots with Seaborn")
plt.grid(True)
plt.show()
```
