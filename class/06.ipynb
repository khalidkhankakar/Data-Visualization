```python
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
x1 = np.random.rand(50)
y1 = np.random.rand(50)


x2 = np.random.rand(50)
y2 = np.random.rand(50) + 1



x3 = np.random.rand(50)
y3 = np.random.rand(50) + 2

x1
```

```python
plt.scatter(x1, y1, label='Data 1', color='red',marker='o')
plt.scatter(x2, y2, label='Data 2', color='blue',marker='o')
plt.scatter(x3, y3, label='Data 3', color='green',marker='o')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Multiple scatter plot')
plt.legend()
plt.show()
```

```python
x1 = [2,3,4,7,4]
y1 = [4,7,9,3,4]


x2 = [1,7,4,3,5]
y2 = [8,5,4,3,6]



x3 =[4,5,6,7,8]
y3 = [1,2,3,4,5]
```

```python
plt.scatter(x1, y1, label='Data 1', color='purple',marker='o')
plt.scatter(x2, y2, label='Data 2', color='green',marker='^')
plt.scatter(x3, y3, label='Data 3', color='black',marker='+')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Multiple scatter plot')
plt.legend()
plt.show()
```

```python
# joint plot
data = sns.load_dataset('iris')
data.head()
sns.jointplot(x='sepal_length', y='sepal_width', data=data,kind='kde')
# give the title 
# plt.title('Joint plot')
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```
