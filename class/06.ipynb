```python
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
x1 = np.random.rand(50)
y1 = np.random.rand(50)


x2 = np.random.rand(50)
y2 = np.random.rand(50) + 1



x3 = np.random.rand(50)
y3 = np.random.rand(50) + 2

x1
```

```python
plt.scatter(x1, y1, label='Data 1', color='red',marker='o')
plt.scatter(x2, y2, label='Data 2', color='blue',marker='s')
plt.scatter(x3, y3, label='Data 3', color='green',marker='^')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Multiple scatter plot')
plt.legend()
plt.show()
```

```python
# joint plot
data = sns.load_dataset('iris')
data.head()
sns.jointplot(x='sepal_length', y='sepal_width', data=data,kind='kde')
#  give the title 
plt.title('Joint plot')
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```
