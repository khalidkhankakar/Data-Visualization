```python
import matplotlib.pyplot as plt 
import numpy as np
# import seaborn as sns
# import pandas as pd
from scipy.stats import norm
from mpl_toolkits.mplot3d import Axes3D
# library for 3d graph
```

```python


# Generate 1000 values between -6 and 6
x = np.linspace(start=-6, stop=6, num=1000)

# Normal distribution with mean=0, standard deviation=1.5
mean = 0
std_dev = 1.5
y = norm.pdf(x, loc=mean, scale=std_dev)

# Plotting
plt.plot(x, y, label=f'Normal Distribution (μ={mean}, σ={std_dev})', color='blue')
plt.title("Normal Distribution Curve")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid(True)
plt.legend()
plt.show()
```

```python
x = np.linspace(-5, stop=5, num=50)
y = np.linspace(-5, stop=5, num=50)

x, y = np.meshgrid(x, y)

z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
# ax = fig.

ax.plot_surface(x, y, z, cmap='rainbow')
ax.set_xlabel('X axies')
ax.set_ylabel('Y axies')
ax.set_zlabel('Z axies')
ax.set_title('3d graph')
plt.show()
```

```python
x = np.linspace(-5, stop=5, num=50)
y = np.linspace(-5, stop=5, num=50)

x, y = np.meshgrid(x, y)

z = np.sin(np.sqrt(x**3 + y**3))

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
# ax = fig.

ax.plot_surface(x, y, z, cmap='rainbow')
ax.set_xlabel('X axies')
ax.set_ylabel('Y axies')
ax.set_zlabel('Z axies')
ax.set_title('cubic shape')
plt.show()
```

```python
x = np.linspace(-4, stop=4, num=100)
y = np.linspace(-4, stop=4, num=100)

x, y = np.meshgrid(x, y)

z = np.sin(np.sqrt(x**2 + y**2))

plt.contour(x,y,z, levels=20, cmap='rainbow')
plt.colorbar()


plt.xlabel('X axies')
plt.ylabel('Y axies')
plt.title('Contour plot')
plt.show()
```
