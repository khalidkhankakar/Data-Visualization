```python
import matplotlib.pyplot as plt 
import numpy as np
# import seaborn as sns
# import pandas as pd
from scipy.stats import norm
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-6, stop=6, num=1000)

#               mean    variance
y = norm.pdf(x, loc=0, scale=1.5)

plt.plot(x, y, label='Normal Distribution (μ=0, σ=1.5)', color='blue')
plt.title("Normal Distribution Curve")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid(True)
plt.legend()
plt.show()
```

```python
x = np.linspace(-5, stop=5, num=50)
y = np.linspace(-5, stop=5, num=50)

x, y = np.meshgrid(x, y)

z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
# ax = fig.

ax.plot_surface(x, y, z, cmap='rainbow')
ax.set_xlabel('X axies')
ax.set_ylabel('Y axies')
ax.set_zlabel('Z axies')
ax.set_title('3d graph')
plt.show()
```

```python
x = np.linspace(-5, stop=5, num=50)
y = np.linspace(-5, stop=5, num=50)

x, y = np.meshgrid(x, y)

z = np.sin(np.sqrt(x**3 + y**3))

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
# ax = fig.

ax.plot_surface(x, y, z, cmap='rainbow')
ax.set_xlabel('X axies')
ax.set_ylabel('Y axies')
ax.set_zlabel('Z axies')
ax.set_title('cubic shape')
plt.show()
```
