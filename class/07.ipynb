```python
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
data = [3,4,2,5,6,2,3,5,1,1]  

# Create by matplotlib
plt.boxplot(data, notch=True)
plt.ylabel("Y label")
plt.xlabel("X label")
plt.title("Box Plot")
plt.show()

# Create by Seaborn
sns.boxplot(y=data, color='red') 
plt.show()


# multiple box plot
# method 1
# df = pd.DataFrame(np.random.rand(10, 4), columns=list('ABCD'))


# method 2
data1 = np.random.rand(10)
data2 = np.random.rand(10)
data3 = np.random.rand(10)

data = [data1, data2, data3]

sns.boxplot(data)
plt.show()
```
