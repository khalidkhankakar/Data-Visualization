```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load a simple real-world dataset (tips)
tips = sns.load_dataset('tips')

# Draw boxplots for total_bill by day with different colors
plt.figure(figsize=(8, 6))
box = plt.boxplot(
    [tips[tips['day'] == d]['total_bill'] for d in tips['day'].unique()],
    patch_artist=True,
    labels=tips['day'].unique()
)

colors = ['lightblue', 'lightgreen', 'lightcoral', 'khaki']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

plt.title('Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```
