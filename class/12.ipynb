```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# No need to redefine 'data' since it already exists

data = np.random.randn(100)
# Create single box plot with color
box = plt.boxplot(data, sym='+', orientation='vertical', notch=
True, whis=0.5,patch_artist=True,
                  boxprops=dict(facecolor='red'))
                  

# Customize the plot
plt.title('Single Box Plot')
plt.ylabel('Values')

plt.show()
```

```python
box_colors = ['pink', 'red', 'blue']

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Sample data
data1 = np.random.randn(100)
data2 = np.random.randn(100) + 2
data3 = np.random.randn(100) - 2


# Create multiple box plots
data = [data1, data2, data3]

# Create multiple box plots with different colors
bp = ax.boxplot(data, patch_artist=True)

for patch, color in zip(bp['boxes'], box_colors):
    patch.set_facecolor(color)


# Customize the plot
plt.title('Multiple Box Plots')
plt.ylabel('Values')
plt.xticks([1, 2, 3], ['Group 1', 'Group 2', 'Group 3'])

plt.show()
```

```python
# Create sample data
data = {
    'Math': np.random.normal(75, 15, 100),
    'Science': np.random.normal(70, 12, 100),
    'English': np.random.normal(80, 10, 100),
    'History': np.random.normal(73, 14, 100)
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Create box plots - simpler method
df.boxplot(column=['Math', 'Science', 'English', 'History'],
    # grid=True,
    figsize=(10, 6),)

# Add labels and title
plt.title('Student Scores by Subject')
plt.ylabel('Scores')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample data
np.random.seed(42)
data = {
    'Group A': np.random.normal(25, 5, 100),
    'Group B': np.random.normal(30, 8, 100),
    'Group C': np.random.normal(35, 6, 100),
    'Group D': np.random.normal(40, 7, 100)
}
df = pd.DataFrame(data)

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Create the boxplot
boxes = ax.boxplot([df[col] for col in df.columns],
                   labels=df.columns,
                   patch_artist=True)

# Define colors
colors = ['red', 'purple', 'coral', 'orange']

# Color each box
for box, color in zip(boxes['boxes'], colors):
    box.set_facecolor(color)

# Customize plot
ax.set_title('Boxplot of Multiple Groups')
ax.set_xlabel('Groups')
ax.set_ylabel('Values')

plt.show()
```
