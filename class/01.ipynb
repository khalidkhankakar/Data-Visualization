```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

# Create a scatter plot
# Add line to the points
plt.scatter(x, y, s=90, color='red',marker='^')  # s is the size of the points
# plt.plot(x, y, linestyle='-', linewidth=2, color='red')  # Line connecting the points
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()

# Or using Seaborn
sns.scatterplot(x=x, y=y, size=1, color='red', marker='^')
plt.show()
```

## Line chart


```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

# Create a line plot
plt.plot(x, y, linestyle='-.', linewidth=2, color='purple')  # linestyle can be '-', '--', '-.', ':'
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.show()

# Or using Seaborn
sns.lineplot(x=x, y=y, linestyle='-.', linewidth=7, color='purple')
plt.show()
```

## Bar chart


```python

# Sample data
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 5, 20]

bar_colors = ['red', 'blue', 'cyan', 'orange']

# Create a bar chart
plt.bar(categories, values, alpha=0.9, edgecolor='red',color=bar_colors)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart")
plt.show()

# Or using Seaborn
sns.barplot(x=categories, y=values, palette=bar_colors, edgecolor='red')
plt.show()
```

### Grouped Bar chart


```python
import numpy as np

x = np.arange(5)

y1 = [34, 56, 12, 89, 67]
y2 = [12, 56, 78, 45, 90]
y3 = [14, 23, 45, 25, 89]
y4 = [22, 44, 66, 33, 77]

width = 0.2
plt.bar(x+0.4, y4, width, color='blue')
plt.bar(x-0.2, y1, width, color='cyan')
plt.bar(x, y2, width, color='orange')
plt.bar(x+0.2, y3, width, color='green')
plt.xticks(x, ['Team A', 'Team B', 'Team C', 'Team D', 'Team E'])
plt.xlabel("Teams")
plt.ylabel("Scores")
plt.show()
```

## stack chart

```python
import numpy as np

x = np.arange(5)

y1 = np.array([34, 56, 12, 89, 67])
y2 = np.array([12, 56, 78, 45, 90])
y3 = np.array([14, 23, 45, 25, 89])
y4 = np.array([22, 44, 66, 33, 77])
# stacked bar chart
plt.bar(x,y1, color='blue')
plt.bar(x, y2,bottom=y1, color='cyan')
plt.bar(x, y3, bottom=y1+y2, color='orange')
plt.bar(x, y4, bottom=y1+y2+y3, color='green')
plt.xticks(x, ['Team A', 'Team B', 'Team C', 'Team D', 'Team E'])
plt.xlabel("Teams")
plt.ylabel("Scores")
plt.show()
```

## Box plot


```python
# Sample data
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 10]  # 10 is an outlier

# Create a box plot
plt.boxplot(data, vert=True,notch=True,sym='+' )  # Set vert=False for horizontal box plot
plt.ylabel("Data")
plt.title("Box Plot")
plt.show()

# Or using Seaborn
sns.boxplot(y=data,color='red', notch=True)  # Use 'y' for vertical orientation
plt.show()
```
