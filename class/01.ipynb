```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

# Create a scatter plot
plt.scatter(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()
plt.show()

# Or using Seaborn
sns.scatterplot(x=x, y=y)

```

## Line chart

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

# Create a line plot
plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.show()

# Or using Seaborn
sns.lineplot(x=x, y=y)
plt.show()
```

## Bar chart

```python

# Sample data
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 5, 20]

# Create a bar chart
plt.bar(categories, values)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart")
plt.show()

# Or using Seaborn
sns.barplot(x=categories, y=values)
plt.show()
```

## Bar plot

```python
# Sample data
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 10]  # 10 is an outlier

# Create a box plot
plt.boxplot(data)
plt.ylabel("Data")
plt.title("Box Plot")
plt.show()

# Or using Seaborn
sns.boxplot(y=data)  # Use 'y' for vertical orientation
plt.show()
```
