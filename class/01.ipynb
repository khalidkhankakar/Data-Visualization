```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

# Create a scatter plot
# Add line to the points
plt.scatter(x, y, s=50)  # s is the size of the points
# plt.plot(x, y, linestyle='-', linewidth=2, color='red')  # Line connecting the points
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()

# Or using Seaborn
sns.scatterplot(x=x, y=y)
plt.show()
```

## Line chart

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

# Create a line plot
plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.show()

# Or using Seaborn
sns.lineplot(x=x, y=y)
plt.show()
```

## Bar chart

```python

# Sample data
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 5, 20]

bar_colors = ['tab:red', 'tab:blue', 'tab:cyan', 'tab:orange']

# Create a bar chart
plt.bar(categories, values, alpha=0.7, edgecolor='black',color=bar_colors)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart")
plt.show()

# Or using Seaborn
sns.barplot(x=categories, y=values)
plt.show()
```

### Grouped Bar chart

```python
import numpy as np

x = np.arange(5)
y1 = [34, 56, 12, 89, 67]
y2 = [12, 56, 78, 45, 90]
y3 = [14, 23, 45, 25, 89]
width = 0.2

plt.bar(x-0.2, y1, width, color='cyan')
plt.bar(x, y2, width, color='orange')
plt.bar(x+0.2, y3, width, color='green')
plt.xticks(x, ['Team A', 'Team B', 'Team C', 'Team D', 'Team E'])
plt.xlabel("Teams")
plt.ylabel("Scores")
plt.show()
```

## Bar plot

```python
# Sample data
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 10]  # 10 is an outlier

# Create a box plot
plt.boxplot(data)
plt.ylabel("Data")
plt.title("Box Plot")
plt.show()

# Or using Seaborn
sns.boxplot(y=data)  # Use 'y' for vertical orientation
plt.show()
```
